```python
import numpy as np
from datascience import *

# These lines do some fancy plotting magic
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

import seaborn as sns
```

```python
CO2 = Table.read_table("Natural_Science_Dataset_2.csv")
CO2
```

```python
CO2.group("SOURCE").column(0)
```

```python
Annual_per_flight_type = CO2.pivot("SOURCE", "Country", values = "Value", collect = sum)
Annual_per_flight_type
```

```python
Annual_per_flight_type.select(0,1).barh("Country")
```

```python
Cleansed_table = CO2.where(
    "SOURCE", are.contained_in(
        ['NRES_DOM_IN', 'NRES_INT_FROM', 'RES_DOM_IN', 
         'RES_DOM_OUT', 'RES_INT_FROM', 'RES_INT_OUT','RES_INT_TO'])).where(
    "FREQUENCY", "A").where("Flight type", "All flights" ).where("TIME", are.below("2023"))

Overview_sum = Cleansed_table.select("Country","Value").group("Country", sum).sort("Value sum", descending =True)
Overview_sum
```

```python
Cleansed_table.group("Flight type")
```

```python
top_10 = Overview_sum.take(np.arange(10))
top_10.barh("Country")
```

```python
US= Cleansed_table.where("Country","United States").select("Value","TIME").group("TIME", sum)
US
```

```python
for country in top_10.column(0):
    country_table = Cleansed_table.where("Country", country).select("Value","TIME").group("TIME", sum)
    sns.lineplot(data= country_table, x="TIME", y="Value sum", label=country)
    
plt.xlabel('Time')
plt.ylabel('Value Sum')
plt.title('Value over Time for Top 10 Countries')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left') 
#palette = sns.color_palette("viridis", len(top_10.column(0)))
plt.figure(figsize=(12, 8)) 

plt.tight_layout()
plt.show()
```
